## Exploring the Wine Dataset

```python
import numpy as np
import seaborn as sb
import mltools as ml
import matplotlib.pyplot as plt
import pandas as pd
```

```python
red_wine_data = pd.read_csv('data/winequality-red.csv',sep=";")
red_wine_data.head()
```

```python
white_wine_data = pd.read_csv('data/winequality-white.csv',sep=";")
white_wine_data.head()
```

### Shape of features and data points

```python
white_wine = np.genfromtxt('data/winequality-white.csv', delimiter=";", skip_header=1)
X = white_wine[:,0:-1] # All other columns are features, exclude wine quality
Y = white_wine[:,-1] # Target value is last column, wine quality
print(f"Number of data points in white_wine dataset: {X.shape[0]}")
print(f"Number of features in white_wine dataset: {X.shape[1]}")
```

```python
red_wine = np.genfromtxt('data/winequality-red.csv', delimiter=";", skip_header=1)
Xr = red_wine[:,0:-1] # All other columns are features, exclude wine quality
Yr = red_wine[:,-1] # Target value is last column, wine quality
print(f"Number of data points in red_wine dataset: {Xr.shape[0]}")
print(f"Number of features in red_wine dataset: {Xr.shape[1]}")
```

### Means for each column:

```python
for column, mean in zip(white_wine_data, np.mean(X, axis = 0)):
    print(f"{column} mean: {mean}")
```

### Varience for each column:

```python
for column, varience in zip(white_wine_data, np.var(X, axis = 0)):
    print(f"{column} mean: {varience}")
```

### Standard Deviation for each column:

```python
for column, std in zip(white_wine_data, np.std(X, axis = 0)):
    print(f"{column} mean: {std}")
```

### Histogram plot of each feature

```python
white_wine_data.hist(bins=11,figsize=(12,15))
```

### Heatmap of feature correlations

```python
wine_labels = ["fixed_acidity", "volatile_acidity", "citric_acid", "residual_sugar",
               "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide", "density",
               "pH", "sulphates", "alcohol", "wine_quality"]

features_dict = dict()
for col, feature in enumerate(wine_labels):
    features_dict[feature] = white_wine[:,col]
    
features = list(features_dict.values())
correlations=np.corrcoef(features)

sb.set(rc={'figure.figsize':(12,10)})
sb.heatmap(correlations, annot=True, xticklabels = wine_labels, yticklabels = wine_labels)
```

#### Correlation heat map shows that level of wine quality is correlated the most with alcohol.

```python
Xtr,Xte,Ytr,Yte = ml.splitData(X,Y,0.75) # split data 75 percent
print(Xtr.shape, Ytr.shape, Xte.shape, Yte.shape)
```

#### Model Exploration of various hidden layers and nodes


```python
nn = ml.nnet.nnetClassify()
hidden_layers = [1,2,4,6]
nodes = [5,10,25,50,100]
Xtr_err = np.zeros((len(hidden_layers), len(nodes)))
Xte_err = np.zeros((len(hidden_layers), len(nodes)))
for i, h in enumerate(hidden_layers):
    for j, n  in enumerate(nodes):
        nn.init_weights([Xtr.shape[1]]+ [n]*h + [7], "random", Xtr, Ytr)
        nn.train(Xtr, Ytr, stopTol=1e-8, stepsize=0.25, stopIter=300)
        Xtr_err[i][j] = nn.err(Xtr, Ytr)
        Xte_err[i][j] = nn.err(Xte, Yte)


f, ax = plt.subplots(1,1, figsize=(8,5))
cax = ax.matshow(Xtr_err, interpolation="nearest")
f.colorbar(cax)
ax.set_xticklabels([""]+hidden_layers)
ax.set_yticklabels([""]+nodes)
plt.show()

f, ax = plt.subplots(1, 1, figsize=(8,5))
cax = ax.matshow(Xte_err, interpolation="nearest")
f.colorbar(cax)
ax.set_xticklabels([""]+hidden_layers)
ax.set_yticklabels([""]+nodes)
plt.show()
```
